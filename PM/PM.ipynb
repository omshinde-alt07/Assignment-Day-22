Q.1

The Iris dataset is an example of a supervised learning problem because it contains input features (like sepal length, petal length, etc.) along with known output labels (species of flower).

It is mainly a classification problem because the target variable (species) is categorical, consisting of three classes: setosa, versicolor, and virginica. Since we are predicting categories and not continuous values, it is classification. There is no reinforcement learning involved because there is no agent or reward-based system.

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

print(df.isnull().sum())


df = df.dropna()


features = df[['sepal length (cm)', 'petal length (cm)']]
target = df['species']
```

Q.3

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

y_reg = df['sepal length (cm)']
X_reg = df[['petal length (cm)']]

model_reg = LinearRegression()
model_reg.fit(X_reg, y_reg)

pred = model_reg.predict(X_reg)

mse = mean_squared_error(y_reg, pred)
print("MSE:", mse)
```

Q.4

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Prepare classification data
X_clf = df[['sepal length (cm)', 'petal length (cm)']]
y_clf = df['species']

model_clf = LogisticRegression(max_iter=200)
model_clf.fit(X_clf, y_clf)

pred_clf = model_clf.predict(X_clf)

acc = accuracy_score(y_clf, pred_clf)
print("Accuracy:", acc)
```

Feature selection has a strong impact on both regression and classification models. In regression, selecting better and more relevant features helps the model make more accurate predictions and reduces error, such as lowering the Mean Squared Error (MSE). It also removes unnecessary variation in the data, making the model more stable. In classification, feature selection improves accuracy by helping the model clearly distinguish between different classes. It also reduces overfitting, meaning the model performs better on new data instead of just memorizing the training data.

Q1 вЂ” Types of Machine Learning (with examples)
Machine learning is mainly divided into three types: supervised learning, unsupervised learning, and reinforcement learning. In supervised learning, the model is trained using labeled data, meaning both input and output are known. For example, predicting house prices or classifying emails as spam or not spam. In unsupervised learning, the model works with unlabeled data and tries to find patterns or groupings, such as customer segmentation using clustering. Reinforcement learning is different, where an agent learns by interacting with an environment and receives rewards or penalties based on its actions, like training a game-playing AI.

Q3 вЂ” Difference between regression and classification
Regression and classification are both types of supervised learning, but they differ in the type of output they produce. Regression is used when the output is continuous, such as predicting temperature, price, or height. On the other hand, classification is used when the output is categorical, such as predicting whether an email is spam or not or identifying the species of a flower. So, the main difference is that regression predicts numerical values, while classification predicts categories.
