Q1 — Difference between loc and iloc in Pandas
In Pandas, loc and iloc are used for selecting data, but they work differently. The loc function is label-based, which means it selects rows and columns using their names or labels. For example, we can select rows using index labels or filter using conditions. On the other hand, iloc is index-based, meaning it selects data using integer positions (like row number and column number). So, loc is used when we know the labels, while iloc is used when we want to access data by position.

Q2 — Filter rows where a column value is greater than average
To filter rows where a column value is greater than its average, we first calculate the mean of that column and then apply a condition. For example, if we want to filter rows where sepal length is greater than average, we can write:

avg = df['sepal length (cm)'].mean()
filtered_df = df[df['sepal length (cm)'] > avg]

print(filtered_df.head())

Q3 — Purpose of describe() and insights from it
The describe() function in Pandas is used to get a summary of the dataset’s numerical columns. It provides important statistical values like count, mean, standard deviation, minimum, maximum, and quartiles. From this, we can understand the overall distribution of data, check how spread out the values are, and identify any unusual values. For example, the mean tells us the average, standard deviation shows variability, and min/max help us understand the range of the data.
